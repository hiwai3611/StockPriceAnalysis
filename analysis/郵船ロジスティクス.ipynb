```python
import pandas as pd
import numpy as np
%matplotlib inline
from fbprophet import Prophet

df = pd.read_csv("C:/Users/h-iwai/Documents/kabu/20170604/csv/9370_郵船ロジスティクス.csv")
df.tail()
```

```python
df['y'] = np.log(df['y'])
#df.tail()

m = Prophet()
m.fit(df);

future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
