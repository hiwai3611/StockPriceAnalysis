```python
import pandas as pd
import numpy as np
%matplotlib inline
from fbprophet import Prophet
```

```python
df = pd.read_csv("bpad.csv")
```

```python
df.tail()
```

```python
ex_df=pd.read_csv("ex.csv")
ex_df['y'] = np.log(ex_df['y'])
ex_df.head()
```

```python
m = Prophet()
m.fit(ex_df);
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
```

```python
forecast.columns
```

```python
m.plot(forecast);
```

```python
m.plot_components(forecast);
```
